###  5) Top 25% of students based on their math score

```python
data.describe()
```

```python
total_marks
```

```python
total_mark = total_marks.values
```

```python
# Add new column 'Total_marks' to the dataframe and storing it as a new dataframe 'data2'
data2=data.assign(Total_marks = total_mark)
```

```python
data2
```

```python
# extracting the rows having math score > 77 and storing it in a new data frame 'T1' which comes to 231 rows
T1 = data2[data2['math score']>77]
```

```python
T1
```

```python
# extracting 19 rows filtered by 'Total_marks' in descending order,having math score = 77 and storing in a new data frame'T77' 
T77 = data2[data2['math score']==77].sort_values(by='Total_marks', ascending=False).head(19)
```

```python
T77
```

```python
# merging the 2 dataframes 'T1' and 'T77' based on common columns
top_25 = pd.merge(T1, T77, on =['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course', 'math score', 'reading score', 'writing score', 'Total_marks'], how = 'outer')
```

```python
# sorting the dataframe 'top_25' based on 'math score' and 'Total marks' in descending order
top_25 = top_25.sort_values(by=['math score', 'Total_marks'], ascending=False) 
```
